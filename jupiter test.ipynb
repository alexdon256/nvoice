```python
%cd /workspace
!rm -r ./nvoice
!rm -r ./Project
!git clone https://github.com/alexdon256/nvoice.git
!pip install -e ./nvoice
import yt_dlp
import subprocess
def download(link, name='Input', cookies_file='youtube_cookies.txt'): 
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio/best[ext=m4a]/best',
        'output': f'{name}.mp4',
        'format': 'mp4'
    }
    command = ['yt-dlp']
    for key, value in ydl_opts.items():
        command.extend([f'--{key}', str(value)])
    command.extend(['--cookies', cookies_file, link])

    try:
        subprocess.run(command, check=True, capture_output=True, text=True)
        downloaded_file_path = f"{name}.mp4"  # Assuming mp4 as output
        print(f"Видео {downloaded_file_path} успешно загружено!")
        return downloaded_file_path
    except subprocess.CalledProcessError as e:
        print(f"Ошибка при загрузке: {e}")
        print(f"Stdout: {e.stdout}")
        print(f"Stderr: {e.stderr}")
        return None

Source_Language = 'en' # @param {type:"string"};
Destination_Language = 'ru' # @param {type:"string"};
Accent_Language = 'ru' # @param {type:"string"};

!mkdir ./Project
%cd ./Project
import os
Project = os.getcwd()
Result_Video = Project+'/result.mp4'
cookies_file_path = '../nvoice/cookies.txt'
Video = download('https://www.youtube.com/watch?v=dm6ux6d6kCA', 'input',cookies_file_path)
!nvoice {Project} {Video} {Source_Language} {Destination_Language} {Accent_Language}
```
